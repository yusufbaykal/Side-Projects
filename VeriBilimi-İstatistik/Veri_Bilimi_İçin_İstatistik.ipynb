# Web Sitesinde Geçirilen Sürenin Testi

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
olcumler = np.array([17,160,234,149,145,107,197,75,201,225,211,119,157,145,127,244,163,114,145,65,112,185,202,146,203,
                    114,188,156,187,154,177,95,165,50,110,216,138,151,166,135,155,84,251,173,131,207,121,120])
```

```python
olcumler[0:10]
```

```python
import scipy.stats as stats
```

```python
stats.describe(olcumler)
```

```python
#Varsayımlar
#normallik varsayımı
```

```python
#histogram
pd.DataFrame(olcumler).plot.hist();
```

```python
#qqplot
import pylab
stats.probplot(olcumler,dist="norm",plot=pylab)
pylab.show()
```

```python
#Shapiro- Wilks Testi
```

H0: Örnek Dağılım ile teorik normal dağılım arasında istatistiksel olarak anlamlı bir farklılık yoktur.
    H1: Örnek Dağılım ile teorik normal dağılım arasında istatistiksel olarak anlamlı bir farklılık vardır.

```python
from scipy.stats import shapiro
```

```python
shapiro(olcumler)
#Sol tarafta görülen değer test istatistiğini ifade edeiyor.
#Sağ tarafta görülen değer p-value değerini temsil ediyor.
```

```python
print("T İstatistiği:"+ str(shapiro(olcumler)[0]))
print("P-Value Değeri:"+ str(shapiro(olcumler)[1]))
```

```python
#Hipotez testinin uygulanması
```

```python
stats.ttest_1samp(olcumler,popmean = 170)
```

# Bağımsız iki Örneklem Testi (A/B)

H0: M1=M2
H1ç M1!=M2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#VERİLERİN İŞLENMESİ 1
```

```python
A = pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,25,21,18,24,26,33,26,27,28,19,25])
B = pd.DataFrame([37,39,31,31,34,38,30,36,29,28,39,29,37,37,30,32,31,31,27,32,33,33,33,31,32,33,26,32,33,29])

A_B = pd.concat([A,B],axis=1)
A_B.columns = ["A","B"]

A_B.head()
```

```python
#VERİLERİN İŞLENMESİ 2
```

```python
A = pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,25,21,18,24,26,33,26,27,28,19,25])
B = pd.DataFrame([37,39,31,31,34,38,30,36,29,28,39,29,37,37,30,32,31,31,27,32,33,33,33,31,32,33,26,32,33,29])
#A ve A'nın Grubu
GRUP_A = np.arange(len(A))
GRUP_A = pd.DataFrame(GRUP_A)
GRUP_A[:] = "A"
A =pd.concat([A,GRUP_A],axis=1)
#B ve B'nın Grubu
GRUP_B = np.arange(len(B))
GRUP_B = pd.DataFrame(GRUP_B)
GRUP_B[:] = "B"
B = pd.concat([B,GRUP_B],axis=1)
```

```python
#Tüm Veri

AB= pd.concat([A,B])
AB.columns = ["gelir","GRUP"]
print(AB.head())
print(AB.tail())
```

```python
sns.boxplot(x = "GRUP", y = "gelir",data=AB);
```

```python
A_B.head()
```

```python
AB.head()
```

```python
#normallik varsayımı
```

```python
#Sol tarafta görülen değer test istatistiğini ifade edeiyor.
#Sağ tarafta görülen değer p-value değerini temsil ediyor.
shapiro(A_B.A)
```

```python
print("T İstatistiği:"+ str(shapiro(A_B.A)[0]))
print("P-Value Değeri:"+ str(shapiro(A_B.A)[1]))
```

```python
#Varyans homojenliği testi
```

H0: Varyanslar Homojendir,
H1: Varyanslar Homojen Değildir

```python
#Sol tarafta görülen değer test istatistiğini ifade edeiyor.
#Sağ tarafta görülen değer p-value değerini temsil ediyor.
stats.levene(A_B.A,A_B.B)
```

```python
print("T İstatistiği:"+ str(stats.levene(A_B.A,A_B.B)[0]))
print("P-Value Değeri:"+ str(stats.levene(A_B.A,A_B.B)[1]))
```

# VARYANS ANALİZİ

```python
#Varyans Analizi: iki ya da daha fazla grup ortalaması arasında farklılık olup olmadığını öğrenilmek adına yapılır.
```

```python
A = pd.DataFrame([32,25,36,24,28,39,35,26,12,37,28,36,34,21,25,26,27,28,39,34,35])
B = pd.DataFrame([40,41,42,39,38,37,36,35,34,27,30,29,30,34,34,31,35,35,33,30,28])
C = pd.DataFrame([40,41,39,38,37,37,25,26,27,30,31,32,37,29,28,42,42,32,31,30,31])
dfs = [A,B,C]
ABC = pd.concat(dfs,axis =1)
ABC.columns= ["GRUP_A","GRUP_B","GRUP_C"]
ABC.head()
```

```python
#VARSAYIM KONTROLÜ
```

```python
from scipy.stats import shapiro
```

```python
shapiro(ABC["GRUP_A"])
```

```python
shapiro(ABC["GRUP_B"])
```

```python
shapiro(ABC["GRUP_C"])
```

```python
stats.levene(ABC["GRUP_C"],ABC["GRUP_B"],ABC["GRUP_C"])
```

```python
print("T İstatistiği:"+ str(stats.levene(ABC["GRUP_C"],ABC["GRUP_B"],ABC["GRUP_C"])[0]))
print("P-Value Değeri:"+ str(stats.levene(ABC["GRUP_C"],ABC["GRUP_B"],ABC["GRUP_C"])[1]))
```

```python
#Hipotez Testi
```

```python
from scipy.stats import f_oneway
```

```python
f_oneway(ABC["GRUP_C"],ABC["GRUP_B"],ABC["GRUP_C"])
```

```python
#Nonparametrik Hipotez Testi
```

```python
from scipy.stats import kruskal
```

```python
kruskal(ABC["GRUP_C"],ABC["GRUP_B"],ABC["GRUP_C"])
```
