Liste Düzleştirme

```
a =  [[1,'a',['cat'],2],[[[3]],'dog'],4,5]
def duzles(a):
    duz_list = []
    for i in a:
        if type(i) is list:
            duz_list.extend(duzles(i))
        else:
            duz_list.append(i)
    return duz_list
```

```
duzles(a)
```
