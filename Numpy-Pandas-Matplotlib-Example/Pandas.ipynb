```python
import numpy as np
import pandas as pd
m = np.random.randint(1,30, size = (10,3))
df = pd.DataFrame(m, columns = ["var1","var2","var3"])
```

```python
df.var1
```

```python
type(df)
```

```python
# var1 değerleri
df["var1"]
```

```python
#Min,median ve max değerlerinin grup halinde yazdırılması
df.groupby("var2").aggregate([min, np.median, max])
```

```python
df.groupby("var1").aggregate({"var2": 'min', "var3": 'max'})
```

```python

#Titanic datasının yüklenmesi
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#Ortalamalara bakılması
titanic.groupby("sex")[["survived"]].mean()
```

```python
#Cinsiyete göre bireylerin hangi sınıfta yer aldığının kontorlü
titanic.groupby(["sex","class"])[["survived"]].aggregate("count").unstack()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean")
```

```python
titanic.pivot_table("survived",index = "sex",columns="class")
```

```python
titanic.age.head()
```

```python
#Yaşları 18-90 arası kişileri sınıflandırmasını yapmak
age = pd.cut(titanic["age"],[0,18,90])
age.head(10)
```

```python
#Yaşların sınıflandırılmasının cinsiyet faktörüne göre ve First,Second,Third gibi değişkenler arasındaki dağılımını göstermek.
titanic.pivot_table("survived",["sex",age],"class")
```
